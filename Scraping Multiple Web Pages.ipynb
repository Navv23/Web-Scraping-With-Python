```python
#scraping multiple web pages!
from bs4 import BeautifulSoup as bs
import requests 
import pandas as pd 
```

```python
url = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_6_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_6_0_na_na_na&as-pos=6&as-type=TRENDING&suggestionId=mobiles&requestId=4b792a99-1dd9-4294-bd97-3555937d47b3&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3D20000&page=1'
page = requests.get(url)
soup = bs(page.text, 'lxml')
```

```python
df = pd.DataFrame({'link':[],'name':[],'price':[],'rating':[],'rating&review_c':[],'description':[]})
while True:
    
    postings = soup.find_all('div',{'class':'_2kHMtA'})
    for post in postings:
        try:
            link = post.find('a',class_='_1fQZEK').get('href')
            full_link = 'https://www.flipkart.com'+link
            name = post.find('div',class_='_4rR01T').text
            price = post.find('div',class_='_30jeq3 _1_WHN1').text
            rating = post.find('div',class_='_3LWZlK').text
            r_r_c = post.find('span',class_='_2_R_DZ').text
            des_1 = post.find('ul',class_='_1xgFaf').text
            df = df.append({'link':full_link,'name':name,'price':price,'rating':rating,'rating&review_c':r_r_c,'description':des_1},ignore_index=True)
        except:
            pass
    next_page = soup.find('a',class_='_1LKTO3').get('href')
    next_page_full = 'https://www.flipkart.com'+next_page
    url = next_page_full
    page = requests.get(url)
    soup = bs(page.text, 'lxml')
```
